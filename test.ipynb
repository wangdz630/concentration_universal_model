```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib 
# matplotlib
os.environ['PATH'] = '/software/texlive-20240311/bin/x86_64-linux:' + os.environ['PATH']
matplotlib.rc_file('/software/mess_config/beauty/matplotlibrc_dingBeauty_v3.9.3')

# 1loadtxt
data = np.loadtxt('./halo_properties_0.0_LCDM.dat', skiprows=1)
M = data[:,0]
c = data[:,4]

data2 = np.loadtxt('./halo_properties2_0.0_LCDM.dat', skiprows=1)
M2 = data2[:,0]
c2 = data2[:,4]

plt.figure(figsize=(6, 6))
plt.semilogx(M, c, 'k.', markersize=1, alpha=0.5)  
plt.semilogx(M2, c2, 'r.', markersize=1, alpha=0.5)
plt.xlabel('Mvir [M_sun/h]', fontsize=12)
plt.ylabel('c_vir', fontsize=12)
plt.title('Concentration vs Mass (z=0.0, MultiDark, LDM)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.ylim(2,20)
plt.show()
```
